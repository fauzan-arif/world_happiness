```python
top_10 = df.head(11)
top_10
```

```python
bottom_10 = df.tail(10)
bottom_10
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Columns representing factors explaining happiness score
factor_columns = [
    'Explained by: Log GDP per capita',
    'Explained by: Social support',
    'Explained by: Healthy life expectancy',
    'Explained by: Freedom to make life choices',
    'Explained by: Generosity',
    'Explained by: Perceptions of corruption'
]
# Melt the DataFrame to long format
melted_df = pd.melt(top_10, id_vars=['Country name', 'Ladder score'], value_vars=factor_columns, var_name='Factor', value_name='Contribution')

# Plotting
plt.figure(figsize=(18, 8))
sns.barplot(x='Contribution', y='Country name', hue='Factor', data=melted_df, orient='h')

# Customize plot
plt.xlabel('Contribution to Happiness Score')
plt.ylabel('Country')
plt.title('Factors Contribution to Happiness Score for Top 10 Happiest Countries')
plt.legend(title='Factor', loc='lower right', bbox_to_anchor=(1, 0))
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

```python
# Melt the DataFrame to long format
melted_df_bottom10 = pd.melt(bottom_10, id_vars=['Country name', 'Ladder score'], value_vars=factor_columns, var_name='Factor', value_name='Contribution')

# Plotting
plt.figure(figsize=(18, 8))
sns.barplot(x='Contribution', y='Country name', hue='Factor', data=melted_df_bottom10, orient='h')

# Customize plot
plt.xlabel('Contribution to Happiness Score')
plt.ylabel('Country')
plt.title('Factors Contribution to Happiness Score for Bottom 10 Happiest Countries')
plt.legend(title='Factor', loc='lower right', bbox_to_anchor=(1, 0))
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

```python
# Calculate average contribution for each factor
average_contributions = df[factor_columns].mean()

# Sort factors by average contribution
top_3_factors = average_contributions.sort_values(ascending=False).head(3)
bottom_3_factors = average_contributions.sort_values(ascending=True).head(3)

print("Top 3 factors contributing the most towards happiness:")
print(top_3_factors)

print("\nBottom 3 factors contributing the least towards happiness:")
print(bottom_3_factors)
```

```python
# Plotting
plt.figure(figsize=(10, 6))

# Plot top 3 factors
plt.barh(top_3_factors.index, top_3_factors, color='lightblue', label='Top 3 Factors')

# Plot bottom 3 factors
plt.barh(bottom_3_factors.index, bottom_3_factors, color='salmon', label='Bottom 3 Factors')

# Customize plot
plt.xlabel('Average Contribution')
plt.ylabel('Factor')
plt.title('Top 3 and Bottom 3 Factors Contributing Towards Happiness')
plt.legend()

plt.show()
```

### Findings Summary

After visualising the data above I came to the following conclusions, in context of the questions we tried to investigate from this data set:
- Earnings is indeed the top most determinant factor for happiness. In our data set in come is being represented by GDP per capita. We can see the consistent theme in both top 10 countries and bottom 10 countries in the stacked bar plots.
- While we dont have actual corruption numbers but perception of corruption as a more measurable metric. We can clearly see this trend that it is one of the lowest contributing factor in happiness score.
- Social system is the second most important factor in determining the happiness of people in a country. Its evident from the bar plot and also the overall average contribution scores.
- Summary: The combination of earning and robust social system support are the orimary factors in cotributing to the perception of how happy citizens of a particular country feel. There are ofcourse few outliers where countries ravaged by war and chaos have lack of social security systems and they are also in the bottom 10 in happines countries index. One surprising finding from this excersise was that healthy life expectancy is scoring lower than income, social support and freedom to make life choices.
